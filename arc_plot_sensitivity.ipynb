```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

df_rows = pd.read_csv("result/arc-data-size.csv", usecols=["input rows","time"])
df_rules = pd.read_csv("result/arc-rule-sensitivity.csv", usecols=["input rules","time"])


```

```python
ax = df_rows.plot(x="input rows", y="time")
ax.set_ylabel("time [s]")
ax.legend(["arc"])
ax = df_rules.plot(x="input rules", y="time")
ax.set_ylabel("time [s]")
ax.legend(["arc"])
```
