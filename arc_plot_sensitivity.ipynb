```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

df_rows = pd.read_csv("result/arc-data-size.csv")
df_rules = pd.read_csv("result/arc-rule-sensitivity.csv")


```

```python
ax = df_rows.plot(x=["input rows"], y=["time_arc","time_rcba"])
ax.set_ylabel("time [s]")
ax.legend(["arc","rCBA"])
ax = df_rules.plot(x="input rules", y=["time_arc","time_rcba"])
ax.set_ylabel("time [s]")
ax.legend(["arc","rCBA"])
```
